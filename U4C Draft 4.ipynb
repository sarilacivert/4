# Cluster Number Determination and Evaluation

Before we begin on the clustering algorithms, let's determine the optimal number of cluster in our dataset.

# The Elbow Method

```python
sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=10000).fit(dfylds)
    dfylds["clusters"] = kmeans.labels_
    #print(data["clusters"])
    sse[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()), color="red")
plt.xlabel("Cluster Number", color="blue")
plt.ylabel("Sum of Squared Errors", color="blue")
plt.title("Elbow Method", color="blue")
plt.show()
```

The elbow method reveals a cluster number of 3 before the curve flattens horizontally.  Specifically, the sum squared errors falls in number on the vertical axis as the number of clusters increases on the horizontal axis reaching the number datapoints in the dataset of 4,193 in the aforementioned, dfyields.describe( ) further above.     

# First Clustering Algorithm:  K - Means

An iterative algorithm that processes each datapoint to minimize the sum of squared differences similar to supervised learning with regression is the K - Means Clustering Algoritm.  The minimizing cost function is inertia that attempts to chose the means or centroids with a parameter as the number of clusters and is scalable, works well with flat geometry and smaller sets of clusters.  If your dataset knowledge is neither strong or weak as you know the number of clusters, but not much else, then K - Means is a good start to move you to other clustering algorithms.  It helps, that K - Means is straightforward, but it is more of a partitioning algorithm, then a clustering algorithm.     

```python
# K-Means begins by normalization of the data
X_norm = normalize(dfylds)

# Fit the data
X_pca = PCA(9).fit_transform(X_norm)

# Calculate predicted values 
y_pred = KMeans(n_clusters=3, random_state=42).fit_predict(X_pca)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_pred)
plt.show()
```

# Second Clustering Algorithm:  Means Shift

The next algorithm changes the methodology to a parameter bandwith from zero to one, taking away scalability with the no presumption of the data or the number of clusters.  More importantly, we move from flat to non - flat geometry or from the plane geometry of a two dimensional plane, i.e. flat surfaces to spherical geometry.  The number of clusters is low so we change the parameter bandwith to a higher 3.25 with 4,193 datapoints as this clustering algorithm assigns probabilities to datapoints and overshoots the number of clusters to 4.   


```python
# Set the bandwidth
bandwidth = estimate_bandwidth(dfylds, n_samples=4193)

# Declare and fit the model.
ms = MeanShift(bandwidth=3.25, bin_seeding=True)
ms.fit(dfylds)

# Extract cluster assignments for each data point.
labels = ms.labels_

# Coordinates of the cluster centers.
cluster_centers = ms.cluster_centers_

# Count our clusters.
n_clusters_ = len(np.unique(labels))

print("Number of estimated clusters: {}".format(n_clusters_))
```

# Third Clustering Algorithm:  Spectral 

Staying with non-flat geometry, but changing the parameters and scalability relative to the number of clusters is the Spectral Clustering Algorithm.  Inputting the number of clusters to 3, yields a plot of 3 clusters 


```python
# Three clusters of 3 in Fixed Income, 3 in FX and 3 in Equity, also from the Elbow Method
n_clusters=3

# Declare and fit the model.
sc = SpectralClustering(n_clusters=n_clusters)
sc.fit(dfylds)
```

```python
predict=sc.fit_predict(dfylds)

predict

plt.plot (predict)
```

# Fourth Clustering Algorithm:  Affinity Propagation

Like Mean - Shift, the Affinity Propagation is not scalable and works with non - flat geometry, but with parameter dampening.  The number of clusters is not an input, but an output of the clustering algorithm.  Like K-Means the data is partitioned attempting to find different sets that are globular or generalized in higher dimensions.  The scikit does require some changes to the arguments as number of clusters in our dataset is low, just like Means Shift.  The parameter damping has changed from the default of 0.5 to almost one and the 200 defaulted iterations reduced to 90 leaving us with 3 clusters.         

```python
# Declare the model and fit it in one statement.

af = AffinityPropagation(damping=0.99,max_iter=90).fit(dfylds)
print('Done')

# Pull the number of clusters and cluster assignments for each data point.
cluster_centers_indices = af.cluster_centers_indices_
n_clusters_ = len(cluster_centers_indices)
labels = af.labels_

print('Estimated number of clusters: {}'.format(n_clusters_))

plt.plot (cluster_centers_indices)
```

# In Conclusion

What matters in clustering algorithms is your level of knowledge of the dataset, how associated the different datapoints are to each other and your methodology.  The Fixed Income Yield Curve Rates move in tandem as a hard cluster, but FX and Equity vary in association as a soft cluster.  Both K - Means and Spectral require prior knowledge on the number of clusters, while Means Shift and Affinity Propogation output the number of clusters if you have little knowledge about your dataset.  K - Means is a general purpose clustering algorithm for even sized, small set of clusters and restricted to the methodology of flat geometry mathematics.  The rest of the clustering algorithms are dynamic in non-flat geometry mathematics, but Means Shift and Affinity Propogation require many clusters, not few, and  can only improve after some fine tuning with the parameters.  Both K - Means and Spectral can process a small number of clusters as long as they are even in size.  Altogether, more work is required in clustering algorithms, but the requirements are non-flat geometry and uneven cluster size provide better descriptive data characteristics for unsupervised learning in data science.
